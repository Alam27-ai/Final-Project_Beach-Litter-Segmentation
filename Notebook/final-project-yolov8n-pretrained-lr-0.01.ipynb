```python
from ultralytics import YOLO
# Load a pre-trained model 
model = YOLO('yolov8x-seg.pt')

# Train the model 
results = model.train(data='/kaggle/working/data.yaml', epochs=50, batch = 16, imgsz=[736,544], conf = 0.3, device = [0,1] , workers = 8, pretrained = True, optimizer = 'SGD', lr0 = 0.01, momentum = 0.9) # argumen tinggal tambah
```

```python
from ultralytics import YOLO
import cv2

# Load the trained model
model = YOLO('/kaggle/input/dataset-beachlitter-yolov8x-50e-lr001/other/default/1/best-YOLOv8x-50e-lr001.pt')  # ganti dengan path ke model yang sudah dilatih

# Load an image for testing
image_path = '/kaggle/input/dataset-beachlitter-yolo1-test/test/images/003143.jpg'  # ganti dengan path ke gambar yang ingin diuji
image = cv2.imread(image_path)  # Membaca gambar menggunakan OpenCV
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Konversi ke RGB jika diperlukan

# Perform inference
results = model.predict(source=image_path, device=0, imgsz=[736], show_labels = False, show_conf = False, show_boxes = False, show = True, save = True)  # Gunakan device sesuai dengan perangkat Anda
```
